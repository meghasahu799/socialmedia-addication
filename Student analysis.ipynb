```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data Science WorkShop dataset from seaborn
df = pd.read_csv("Students_Social_Media_Addiction.csv")
df.head()
```

```python
# Check dataset shape (rows, columns)
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Check column data types and missing values
df.info()

# Check summary statistics of numerical columns
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python

# Fill missing 'age' with median
df['Age'].fillna(df['Age'].median(), inplace=True)


df['Most_Used_Platform'].fillna(df['Most_Used_Platform'].mode()[0], inplace=True)
df
```

```python
# Boxplot to check outliers in 'age'
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.show()
```

```python
# Countplot for survival
sns.countplot(x='Country', data=df)
plt.title("Survival Count")
plt.show()
```
